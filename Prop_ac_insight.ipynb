```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
filepath = r'/content/Prop set 1 data - Copy.xlsx'
data = pd.read_excel(filepath)  # Replace 'your_dataset.csv' with your dataset file

# List of variables to create line charts for
variables_to_order = ['Pitch angle', 'Attack angle', 'Tip length']

# Create a line chart for each variable
for variable_to_order in variables_to_order:
    # Sort the data by the chosen variable
    data_sorted = data.sort_values(by=variable_to_order)

    # Extract the sorted variables
    ordered_variable = data_sorted[variable_to_order]
    magnitude = data_sorted['Magnitude']

    # Create a line chart to visualize the gradual change
    plt.figure(figsize=(8, 4))

    plt.plot(ordered_variable, magnitude, marker='o')
    plt.xlabel(variable_to_order)
    plt.ylabel('Magnitude')
    plt.title(f'Magnitude vs. {variable_to_order}')

    plt.show()
```
